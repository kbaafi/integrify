# Tidy Data
Read "Tidy Data" paper by Hadley Wickham

Rules of Tidy Data
* Columns represent separate variables
* Rows represent individual observations
* Observational units form a table

### Reshaping data using melt

```python
import pandas as pd
```

```python
df = pd.read_csv("datasets/airquality.csv")
```

```python
df.head()
```

```python
airquality_melt = pd.melt(frame = df, id_vars=['Month','Day'],
                         value_vars=['Ozone','Solar.R','Wind','Temp'],
                         var_name='measurement', value_name='reading')
airquality_melt.head()
```

# Pivot: unmelting data

* Opposite of melting
* Pivoting turns unique values into separate columns
* Pivots are reporting friendly, melts are analysis friendly
* Pivoting violates tiday data principle

![pivot](images/pivot.png)

![pdpivot](images/pdpivot.png)

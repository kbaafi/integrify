## GroupBy

```python
import pandas as pd

drinks = pd.read_csv("drinks.txt")
drinks.head(20)
```

### Which continent drinks more beer on average

```python
grouped = drinks.groupby(['continent'])['continent','beer_servings'].mean()
max_beer = grouped[grouped.beer_servings==grouped.beer_servings.max()]
max_beer
```

### Statistics on wine consumption

```python
drinks.groupby(['continent'])['continent','wine_servings'].describe()
```

### Mean alcohol consumption per continent

```python
grouped = drinks.groupby(['continent'])['continent',
                                        'beer_servings',
                                        'wine_servings',
                                        'spirit_servings',
                                        'total_litres_of_pure_alcohol'].mean()
grouped
```

### Median alcohol consumption per continent

```python
grouped = drinks.groupby(['continent'])['continent',
                                        'beer_servings',
                                        'wine_servings',
                                        'spirit_servings',
                                        'total_litres_of_pure_alcohol'].median()
grouped
```
